```python
data = []
for url in urls:
    response1 = session.get(url)
    # do something with the webpage, e.g. scrape data
    print(f"Scraping data from {url}")
    soup = bs(response1.html.html, 'html.parser')

    # Link
    link = url


    # Title
    title = soup.find('span', attrs={'id':'productTitle'}).text


    # Subtitle
    try:
        subtitle = soup.find('span', attrs={'id':'productSubtitle'}).text
    except AttributeError:
        subtitle = ''


    # Description 
    des = soup.find('div', attrs={'id':'bookDescription_feature_div'})

    des1 = soup.find('div', attrs={'class':'a-expander-content a-expander-partial-collapse-content'})
    try:
        description = des1.find('span', attrs={'class':'a-text-bold'}).text
    except AttributeError:
        description = ''

    # Rank
    c = soup.findAll('div', attrs={'class':'p13n-amazon-charts-badge-wrapper badge-wrapper'})
    rnk = soup.find('a', attrs={'class':'badge-link'})
    rank = soup.find('span', class_='cat-name').text.strip()

    # Pub date
    import re

    text = subtitle
    match = re.search(r'\b\w+\s\d{1,2},\s\d{4}\b', text)

    if match:
        pub_date = match.group()
    pub_date


    # Author
    d = soup.findAll('div', attrs={'id':'bylineInfo_feature_div'})
    author_span = soup.find('span', class_='author')
    author_name = author_span.a.get_text()
    
    


    data.append({
        'Title': title,
        'Sub_Title': subtitle,
        'Author': author_name,
        'Description': description,
        'Link': url,
        'Date_Published': pub_date,
        'Rank': rank,
    })



    
    if url == urls[-1]:
        break

import csv
# Write the data to a CSV file
with open('data2.csv', 'w', newline='') as csvfile:
    fieldnames = ['Title', 'Sub_Title', 'Author', 'Description', 'Link', 'Date_Published', 'Rank']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()
    for row in data:
        writer.writerow(row)
```
