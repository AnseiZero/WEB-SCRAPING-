```python
import requests
import requests_html
from requests_html import HTMLSession
from bs4 import BeautifulSoup as bs
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import io
```

```python
product_name = []
product_price = []
product_rating = []
product_img_url = []
old_price = []
product_sku = []
```

```python
page = 1
while page != 51:
      url = f"https://www.jumia.com.ng/catalog/?q=laptop&page={page}"
      session = HTMLSession()
      response = session.get(url)
      soup = bs(response.html.html, 'html.parser')
      
      for tag in soup.findAll(attrs={'class':'prd _fb col c-prd'}):
        
        # finding the product name
        div1 = tag.find('div', attrs={'class':'info'})
        name = div1.find('h3', attrs={'class':'name'})
          
        # finding the product price (former price and new price)
        nprice = div1.find('div', attrs={'class':'prc'})
        fprice = div1.find('div', attrs={'class':'old'})
            
        # finding the product SKU
        skus = tag.find('a', attrs={'class':'core'}).get('data-id')
          
        # finding the rating
        rating = div1.find('div', attrs={'class':'stars _s'})
        
        # finding the image url
        img_url = tag.find('img')['data-src']
          
        product_name.append(name.text)
        product_price.append(nprice.text)
        old_price.append(fprice)  
        product_rating.append(rating)
        product_img_url.append(img_url)
        product_sku.append(skus)
      
      page += 1
```

```python
print(old_price[0], product_name[0], product_price[0])
```

```python
product_list = pd.DataFrame({'product_name': product_name,
                             'product_price': product_price,
                             'product_rating': product_rating,
                             'product_img_url': product_img_url,
                             'product_SKU': product_sku,
                             'old_price': old_price})
```

```python
product_list.product_img_url
```

```python
product_list.to_csv('products.csv', index=False, encoding='utf-8')
```
