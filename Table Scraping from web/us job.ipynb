```python
import requests
import requests_html
from requests_html import HTMLSession
from bs4 import BeautifulSoup as bs
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import io
```

```python
url = f"https://www.prizeinfo.net/all-prize-bond-first-second-record/"
session = HTMLSession()
response = session.get(url)
soup = bs(response.html.html, 'html.parser')
      
```

```python
table = soup.find('table', attrs={'class':'table'})
```

```python
data = pd.read_html(str(table))
```

```python
dfs = pd.concat(data)
```

```python
dfs.head(50)
```

```python
# replace NaN with empty string
df = dfs.replace(np.nan, '')
df.head(50)
```

```python
df.info()
```

```python
df.to_csv('prizeinfo.csv',          # Export pandas DataFrame as CSV
            header = True)
```
